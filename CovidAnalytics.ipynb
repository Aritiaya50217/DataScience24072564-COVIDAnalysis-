```python
# รายงานสถานการณ์ COVID-19 ระลอก 3 (ตั้งแต่ 01/04/2021 – 24/08/2021) แยกตามรายจังหวัด

import pandas as pd
import json
from datetime import datetime
import numpy as np

# ดึงข้อมูล การติดเชื้อประจำวันที่ 24 สค 64
with open('timeline-cases-by-provinces.json') as json_file : 
    # json.load() คือการแปลง JSON ไปเป็น Python Object (Dict) 
    data = json.load(json_file)
df = pd.DataFrame(data)
df
```

```python
# ดึงข้อมูลของจังหวัดลพบุรีออกมาทั้งหมด
grouped_1 = df['new_case'] >= 0
grouped_2 = df['province'] == 'ลพบุรี'
lopburi_all = df[(grouped_1) & (grouped_2)]
lopburi_all
```
